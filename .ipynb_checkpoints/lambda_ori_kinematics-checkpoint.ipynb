```python
# Fitting the pm shifts

import scipy as sp

shift_pmra = dx - pmra.value
shift_pmdec = dy - pmdec.value
slope1, intercept1, r_value1, p_value1, std1 = sp.stats.linregress(x,shift_pmra)
slope2, intercept2, r_value2, p_value2, std2 = sp.stats.linregress(distance,shift_pmdec)

f_pmra = slope1*x + intercept1
f_pmdec = slope2*distance.value + intercept2

plt.figure()
plt.scatter(x, shift_pmra)
plt.plot(x,slope1*x + intercept1, c = 'orange')
plt.show()

plt.figure()
plt.scatter(distance, shift_pmdec)
plt.plot(distance,f_pmdec, c = 'orange')
plt.show()



```

```python
def pm_shift_calc(dependence,slope,intercept):
    return dependence*slope + intercept


x_whole = l_ori.ra.values
y_whole = l_ori.dec.values
d_whole = 1000/l_ori.parallax.values
pmra_whole = l_ori.pmra.values
pmdec_whole = l_ori.pmdec.values
pmra_shift_inf = pm_shift_calc(x_whole,slope1,intercept1)
pmdec_shift_inf = pm_shift_calc(d_whole,slope2,intercept2)


plt.figure()
plt.plot(x_whole,pmra_shift_inf)
plt.scatter(x,shift_pmra,c='orange')
plt.show()

plt.figure()
plt.plot(d_whole,pmdec_shift_inf)
plt.scatter(distance,shift_pmdec,c='orange')
plt.show()

pmra_whole = pmra_whole - pmra_shift_inf
pmra_whole = pmra_whole - np.mean(pmra_whole)
pmdec_whole = pmdec_whole + pmdec_shift_inf
pmdec_whole = pmdec_whole - np.mean(pmdec_whole)
```
